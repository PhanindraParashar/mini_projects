```python
L = [2,3,5,7]
def gcd_list(prod):
    gcd_list = []
    for i in range(2,prod):
        if np.gcd(prod,i) == 1:
            gcd_list.append(i)
    return gcd_list
    
def modulo_test_list(L):
    prod = 1
    for i in L:
        prod = prod*i
    gl =  gcd_list(prod)
    return gl

def mod_list(L):
    prod = 1
    for i in L:
        prod = prod*i
    return len(gcd_list(prod))/prod

def mod_prod(prod):
    return len(gcd_list(prod))/prod
```

```python
def isprime(n):
    if n in [2,3,5,7,11]:
        return True
    if n in [0,1,4,6,8,10]:
        return False
    if n > 10 :
        if n  % 6 in [2,3,4,0]:
            return False
        else:
            l = int(np.sqrt(n)+1)
            flag = 1
            for i in range(3,l):
                if n % i == 0:
                    flag = 0
                    break
            if flag == 0:
                return False
            else:
                return True
def primes(n):
    L = []
    for i in range(1,n+1):
        if isprime(i):
            L.append(i)
    return L
```
