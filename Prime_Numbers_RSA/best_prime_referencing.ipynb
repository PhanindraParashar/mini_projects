```python
import numpy as np
```

```python
def isprime(n):
    l = int(np.sqrt(n))
    if n in [1,4,6,8,9,10]:
        return False
    if n in [2,3,5,7] :
        return True
    if n > 10:
        if n%6 in [2,3,4,0]:
            return False
        else:
            flag = 1
            for i in range(3,l+1,2):
                if n % i == 0:
                    flag = 0
            if flag == 0:
                return False
            
            else:
                return True
        
```

```python
def list_primes(n):
    P = []
    for i in range(n):
        if isprime(i):
            P.append(i)
    return P
```

```python
len(list_primes(1000))
```

```python
def no_prime_check(n):
    sum = 0
    L = []
    for i in range(1,n*5):
        if isprime(n+i):
            if (i%n) not in L:
                L.append(i%n)
                sum += 1
    return sum,L
```

```python
def calculate_eff(n):
    return no_prime_check(n)[0]/n
```
