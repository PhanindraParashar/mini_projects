```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()

M = 1000

Nodes = [1,2,3,4,5]
Demand = [-20,-10,11,2,17]
Node_Demand = dict(zip(Nodes,Demand))

Directed_Edges = [(1,3),(1,4),(1,5),(2,3),(2,4),(2,5)]
Cost_Capacity_Edges = [(1,M),(6,M),(3,M),(7,M),(3,M),(1,M)]

Directed_Edges_Cost_Capacity = dict(zip(Directed_Edges,Cost_Capacity_Edges))
```

```python
Directed_Edges_Cost_Capacity
```

```python
for i in Nodes:
    G.add_node(i, demand = Node_Demand[i])

for k in Directed_Edges:
    i,j = k
    cost,cap = Directed_Edges_Cost_Capacity[k]
    G.add_edge(i, j, weight = cost, capacity = cap)
    
flowDict = nx.min_cost_flow(G)
nx.draw_shell(G, with_labels = True)
plt.show()
```

```python
G.add_node('1', demand = -20)
G.add_node('2', demand = -10)
G.add_node('3', demand = 11)
G.add_node('4', demand = 2)
G.add_node('5', demand = 17)

G.add_edge('1', '3', weight = 1, capacity = 1000)
G.add_edge('1', '4', weight = 6, capacity = 1000)
G.add_edge('1', '5', weight = 3, capacity = 1000)
G.add_edge('2', '3', weight = 7, capacity = 1000)
G.add_edge('2', '4', weight = 3, capacity = 1000)
G.add_edge('2', '5', weight = 1, capacity = 1000)


flowDict = nx.min_cost_flow(G)
```
